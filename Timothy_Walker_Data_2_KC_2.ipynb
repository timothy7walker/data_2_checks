```python
import pandas as pd
import numpy as np
```

Data retrieved from [Louisville Metro Data](https://data.louisvilleky.gov/datasets/LOJIC::louisville-ky-metro-parks-2/about).

```python
louisvilleParks = pd.read_csv('Louisville_KY_Metro_Parks.csv')
print(louisvilleParks.shape)
louisvilleParks
```

Removed three columns of data I did not want to include.

```python
louisvilleParks = louisvilleParks.drop(columns=['REGION', 'PUBLIC_', 'CONFERENCE']) 
```

Read column headers for clarity.

```python
louisvilleParks.columns
```

Created an updated list of new column header titles.

```python
fixed_columns = {'OBJECTID' : 'List_number', 'PARK_NAME' : 'Park_Name', 'PARK_ACRE' : 'Park Size (acres)', 'OPS_DIST' : 'District Operating',
                'PARK_TYPE' : 'Park_Type', 'OLMSTED' : 'Olmstead Designed', 'PARK_WEB' : 'Park Web Address', 'FULLADDR' : 'Park Address',
                'RESTROOM' : 'Restroom Avail', 'PICNIC' : 'Picnic Area', 'CAMPING' : 'Camping Area', 'SWIMMING' : 'Swimming', 'HIKING' : 'Hiking',
                'FISHING' : 'Fishing', 'PLAYGROUND' : 'Playground', 'GOLF' : 'Golf', 'SOCCER' : 'Soccer', 'BASEBALL' : 'Baseball', 'BASKETBALL' : 'Basketball',
                'FOOTBALL' : 'Football', 'TENNIS' : 'Tennis', 'HORSEBACK' : 'Horseback', 'VOLLEYBALL' : 'Volleyball', 'WALKING' : 'Walking', 
                'MULTIFIELD' : 'Multifield', 'SPRAYPADS' : 'Spraypads', 'SPRAYGRND' : 'Spraygrounds', 'COMMUNITY' : 'Community Center', 'SPECIAL' : 'Special Center',
                'SKATEPARK' : 'Skatepark', 'SHAPEAREA' : 'Park Area', 'SHAPELEN' : 'Park Length'}
```

Update to column header titles.

```python
louisvilleParks.rename(columns=fixed_columns, inplace=True)
louisvilleParks.columns
```

```python
louisvilleParks.fillna(' ')
```

Removed all 'NaN' data entries and replaced with 'NO'. All 'NaN' entries are the entries meaning not available (No).

```python
 
louisvilleParks = louisvilleParks.replace({'Restroom Avail' : ' ', 'Picnic Area' : ' ', 'Volleyball' : ' ', 'Walking' : ' ', 
                         'Multifield' : ' ',  'Spraypads' : ' ', 'Spraygrounds' : ' ', 'Community Center' : ' ', 
                         'Special Center' : ' ', 'Skatepark' : ' ',}, {'Restroom Avail' : 'NO', 'Picnic Area' : 'NO', 
                         'Volleyball' : 'NO', 'Walking' : 'NO', 'Multifield' : 'NO', 'Spraypads' : 'NO', 'Spraygrounds' : 'NO',
                         'Community Center' : 'NO', 'Special Center' : 'NO', 'Skatepark' : 'NO'},  regex=True)
```
